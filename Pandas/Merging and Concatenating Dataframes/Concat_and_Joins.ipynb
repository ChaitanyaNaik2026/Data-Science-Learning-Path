#Merge

```python
df5 = pd.DataFrame({'id':['A','B','C','D'],'age':[10,20,30,40]})
df6 = pd.DataFrame({'id':['B','C','D','E'],'job':['doctor','plumber','developer','analyst']})
```

```python
df5
```

```python
df6
```

##Inner Join

```python
#inner join
df_inner_join=df5.merge(df6,on ='id',how='inner')
```

```python
print(df5.shape)
print(df6.shape)
print(df_inner_join.shape)
```

```python
#make a copy of df_movie dataframe
df_movies2= df_movie.copy()
```

```python
#set first 1000 values of 'imdb_title_id' column as 'tt1234567890'
for i, index in enumerate(df_movies2.index):
  if i < 1000:
    df_movies2.loc[index,'imdb_title_id']='tt1234567890'
```

```python
df_movies2
```

```python
#merge df_movie with df_rating - left exclusive
df_exclusive_left = df_movies2.merge(df_rating,
                                      left_index=True,
                                      right_index=True,
                                      how='outer',
                                      indicator=True).query("_merge=='left_only'")
```
