##Dataset overview and making pivot table

###Making pivot table

```python
#drop null values
df_population_raw.dropna(inplace=True)
```

```python
#pivot table
df_population = df_population_raw.pivot_table(index='year', columns='country', values='population')
```

```python
df_population
```

```python
#select some countries
df_population = df_population[['United States', 'India', 'China', 'Indonesia', 'Brazil']]
```

```python
df_population
```

##Visualizations

###Lineplot

```python
#4 arguments
df_population.plot(kind='line',xlabel='Year',ylabel='Population',title='Population(1955-2020)',figsize=(8,4))
```

###Barplot

####Single Barplot

```python
#Selecting single country
df_pivot_2020 =df_population [df_population.index.isin([2020])]
```

```python
df_pivot_2020
```

```python
#transpose dataframe
df_pivot_2020 = df_pivot_2020.T
```

```python
df_pivot_2020.plot(kind='bar',color='black',xlabel='Country',ylabel='Population',title='Population in 2020',figsize=(8,4))
```

####Barplot grouped by n variables

```python
#select few year
df_pivot_sample =df_population [df_population.index.isin([1980,1990,2000,2010,2020])]
```

```python
df_pivot_sample
```

```python
df_pivot_sample.plot(kind='bar',xlabel='Year',ylabel='Population')
```

###Piechart

```python
#change column name
df_pivot_2020.rename(columns={2020:'2020'},inplace=True)
```

```python
df_pivot_2020.plot(kind='pie',y='2020',title ='Population in 2020',figsize=(8,4))
```
