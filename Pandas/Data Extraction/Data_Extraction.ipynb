#Filter dataframe with the query() method

##query()

```python
#select player older than 34
df.query('age>34')
#or using boolean slicing
df[df['age']>34]
```

```python
#select player oldet than 34 from italy
df.query('age>34'and'nationality=="Italy"') #Use and for query

#Boolean slicing
df[(df['age']>34) & (df['nationality']=='Italy')]
```

```python
#not operator
df.query('not nationality=="Italy"')
```

```python
#COnvert height to meters and select thove above 1.8
df[df['height_cm']/100 >1.8]
```

```python
df['dob'] = df['dob'].astype('datetime64[ns]')
df['dob'].dt.year
```

#Lambda+apply() method


##Lambda function

```python

def sum_values(a,b):
  x=a+b
  return x
```

```python
sum_values(4,5)
```

```python
#using lambda
sum_values = lambda a,b:a+b
sum_values(4,3)
```
