```python
#Recursion
#Function calling itself
def factorial(n):
    if n == 1:        # Base case
        return 1
    else:
        return n * factorial(n-1)   # Recursive case

print(factorial(4))   # Output: 24
```

```python
#Operator overloading
class X :
   def __init__(self,name,age):
      self.name =name
      self.age = age

   def __gt__(self,other):
      return True if self.age > other.age else False
Alice = X('Alice',7)
bob = X('Bob',8)

print(Alice>bob)
```
